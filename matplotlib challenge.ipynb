## Methods to Improve the ConMediaUse vs ImmiPhobia Scatterplot

Several design improvements can enhance both the clarity and interpretability of the scatterplot displaying the relationship between *ConMediaUse* and *ImmiPhobia*.

First, appropriate **color selection and visual aesthetics** should be applied. Extremely bright or distracting colors can make the figure harder to interpret and distract from the underlying message. Using a balanced and perceptually appropriate color palette improves readability and supports effective communication of the data (Wilke, *Introduction*, p. 1).

Second, the plot should include **clear axis labels, titles, and scales** to ensure the visualization accurately conveys the data and does not mislead the viewer. A visualization must faithfully represent numeric relationships and avoid distortions that could cause viewers to misinterpret differences between values (Wilke, *Introduction*, p. 1).

Third, reducing **visual clutter**—such as overly strong gridlines, inconsistent font sizes, or excessive decorative elements—can improve interpretability. Clean design elements allow viewers to focus on the relationship between variables rather than on distracting stylistic features (Wilke, *Introduction*, p. 1).

Finally, improving **symbol size and transparency** (e.g., smaller markers with partial transparency) can help reduce overplotting and reveal underlying patterns in dense scatterplots, improving the viewer’s ability to detect trends and relationships.

Overall, effective scatterplot improvement requires balancing **accurate scientific representation with aesthetic clarity**, ensuring that the visualization communicates the data truthfully while remaining visually accessible (Wilke, *Introduction*, p. 1).
