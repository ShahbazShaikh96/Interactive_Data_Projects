```python
## Loading the libraries REQUIRED for the project

import pandas as pd
import matplotlib.pyplot as plt

# Loading the dataset
cces = pd.read_csv("cces_example_data.csv")

# Explicitly converting categorical variables (assignment instruction)
categorical_cols = ["sex", "pid3", "nonwhite", "Evaluation"]
for col in categorical_cols:
    cces[col] = cces[col].astype("category")
```

```python
order = ["Poor", "Fair", "Good", "Excellent"]
counts = cces["Evaluation"].value_counts().reindex(order)
```

```python
evaluation_counts = cces["Evaluation"].value_counts()

fig2, ax2 = plt.subplots(figsize=(7,4))
ax2.bar(evaluation_counts.index.astype(str), evaluation_counts.values)

ax2.set_title("Evaluation Frequency")
ax2.set_xlabel("Evaluation Category")
ax2.set_ylabel("Count")

plt.show()
```

```python
## Creating a scatter plot that shows the relationship between "ConMediaUse" and "ImmiPhobia"

# Gridlines help estimate approximate values
# But heavy grids reduce clarity
# Subtle grids should be used only when needed

# Light gridlines were added to help viewers estimate relative values along both axes, improving interpretability without distracting from the plotted data (Wilke, Ch.18).

fig3, ax3 = plt.subplots(figsize=(6,5))

ax3.scatter(cces["ConMediaUse"], cces["ImmiPhobia"], alpha=0.4, s=20)

ax3.set_title("Relationship Between Conservative Media Use and Immigration Attitudes")
ax3.set_xlabel("Conservative Media Use")
ax3.set_ylabel("Immigration Attitudes")

plt.show()
```

## Improving the ConMediaUse vs ImmiPhobia Scatterplot

Several design improvements can enhance the clarity and interpretability of the scatterplot.

First, reducing marker size and applying partial transparency helps mitigate overplotting, allowing viewers to better perceive density patterns and trends within the data (Wilke, 2019, Chapter 18, pp. 184â€“185).

Second, clear axis labels and descriptive titles ensure the visualization accurately communicates the variables being displayed and prevents potential misinterpretation (Wilke, 2019, Chapter 18, p. 186).

Third, minimizing unnecessary decorative elements, such as overly bright colors or distracting graphical components, helps maintain visual focus on the relationship between variables and supports effective figure design (Wilke, 2019, Chapter 18, pp. 188â€“189).

Together, these design principles improve perceptual clarity while ensuring that the visualization remains both accurate and visually accessible (Wilke, 2019, Chapter 18, p. 190).

**Reference**

Wilke, C. O. (2019). *Fundamentals of Data Visualization*. Oâ€™Reilly Media.


```python
## Creating a stacked bar chart that shows the relationship between "pid3" and "Evaluation"

colors = ["#FF6B6B", "#FFD93D", "#6BCB77", "#4D96FF", "#9D4EDD"]

cross_tab = pd.crosstab(cces["pid3"], cces["Evaluation"])

fig4, ax4 = plt.subplots(figsize=(8,5))
cross_tab.plot(kind="bar", stacked=True, ax=ax4, color=colors)

ax4.set_title("Evaluation by Political Identity")
ax4.set_xlabel("Political Identity (pid3)")
ax4.set_ylabel("Count")
ax4.legend(title="Evaluation")

plt.show()
```
