```python
## Loading the necessary libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
import numpy as np
```

```python
## Loading the dataset

Spotify_data = pd.read_csv('spotify-2023.csv', encoding='latin-1')
```

```python
## 1. Create a histogram of the 'bpm' column with a KDE overlay to visualize the distribution of beats per minute in the dataset.
hist1 = sns.histplot(data=Spotify_data, x='bpm', kde=True)
hist1.plot()
```

```python
## 2. Create a bar plot showing the count of tracks released in each month. Use the 'released_month' column for the x-axis and the count of tracks for the y-axis.

releasedmonthgroup = Spotify_data.groupby('released_month').count()["track_name"]
releasedmonthgroup
```

```python
releasemonthvalues = releasedmonthgroup.values
releasemonthvalues
```

```python
releasemonthindex = releasedmonthgroup.index
releasemonthindex
```

```python
## 3. Convert the 'released_month' column to a categorical variable and rename the categories to the corresponding month names (e.g., 1 to 'January', 2 to 'February', etc.). Then, create a categorical plot (catplot) to visualize the count of tracks released in each month.

cat2 = sns.catplot(x = releasedmonthgroup.index, y = releasedmonthgroup.values)
plt.show()
```

```python
## Renaming the categories to the corresponding month names

cat2 = sns.catplot(x = releasedmonthgroup.index, y = releasedmonthgroup.values, kind="bar")
plt.show()
```

```python
## 4. Create a bar plot showing the count of tracks released in each month, with the bars colored based on the count of tracks (use the 'hue' parameter in seaborn). This will help visualize which months had more releases.

cat3 = sns.catplot(x = releasedmonthgroup.index, y = releasedmonthgroup.values, kind="bar", hue = releasedmonthgroup.values)
plt.show()
```

```python
## 5. Create a bar plot showing the count of tracks released in each month, with the bars colored based on the count of tracks (use the 'hue' parameter in seaborn) and a custom color palette (e.g., "Spectral"). This will enhance the visual appeal of the plot while still conveying the information effectively.

cat4 = sns.catplot(x = releasedmonthgroup.index, y = releasedmonthgroup.values, kind="bar", hue = releasedmonthgroup.values, palette="Spectral", legend = False)
plt.show()
```

```python
## 6. Create a scatter plot to visualize the relationship between 'energy_%' and 'bpm' (beats per minute) in the dataset. Use different colors for the points based on the 'released_month' column to see if there are any patterns in energy and bpm across different months.

rel1 = sns.relplot(data = Spotify_data, x = 'energy_%', y = 'bpm')
plt.show()
```

```python
Spotify_data["month_cat"] = Spotify_data["released_month"].astype("category")
Spotify_data["month_cat"] = Spotify_data["month_cat"].cat.rename_categories(releasedmonthindex_names)
```

```python
rel2 = sns.relplot(
    data=Spotify_data,
    x='energy_%',
    y='bpm',
    hue="month_cat",
    palette=cp,
    style='mode',
    markers=["8", ">"],
    s=150   # <-- Increase size
)

plt.show()
```

```python
## 7. Create a scatter plot to visualize the relationship between 'energy_%' and 'bpm', with separate plots for each mode (major and minor). Use the 'col' parameter in seaborn to create separate columns for each mode, and color the points based on the mode as well.

rel3 = sns.relplot(
    data=Spotify_data,
    x='energy_%',
    y='bpm',
    col="mode",
    hue="mode",
    legend=False,
    alpha=0.7,
    s=150
)
plt.show()
```

```python
## 8. Create a scatter plot to visualize the relationship between 'energy_%' and 'bpm', with separate plots for each month of release. Use the 'col' parameter in seaborn to create separate columns for each month, and color the points based on the month as well.

rel4 = sns.relplot(
    data=Spotify_data,
    x='energy_%',
    y='bpm',
    col="released_month",
    col_wrap=4,
    legend=False,
    alpha=0.8,
    s=80
    )
plt.show()
```
